```python
Features= Data_Encoder.iloc[:,:-1]
Features.head()
```

```python
Target= Data_Encoder['status']
Target.head()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
RS = RobustScaler()
```

```python
X_Data=RS.fit_transform(Features)
X_Data
```

### Without Train, Test and split

```python
from sklearn.linear_model import LogisticRegression
```

```python
Lr= LogisticRegression()
```

```python
Lr.fit(X_Data,Target)
```

```python
Lr.score(X_Data,Target)
```

### With Train, Test and split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(Features,Target)
```

```python
X_train.shape, X_test.shape, Y_train.shape,Y_test.shape
```

```python
Lr.fit(X_train,Y_train)
```

```python
Lr_Train_score=Lr.score(X_train,Y_train)
Lr_Train_score
```

```python
Lr_Test_score=Lr.score(X_test,Y_test)
Lr_Test_score
```

```python
Lr.coef_
```

```python
Lr_New_Predict=Lr.predict(X_test)
Lr_New_Predict
```

```python
df=pd.DataFrame({'Actual': Y_test, 'New_Prediction': Lr_New_Predict, 'Error': Y_test- Lr_New_Predict})
df.head()
```

### Model Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm= confusion_matrix(Y_test,Lr_New_Predict)
cm
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(Y_test, Lr_New_Predict))
```

```python
kernel=['linear', 'poly', 'rbf', 'sigmoid']
```

```python
for i in kernel:
    print("Right now my kernel is {}".format(i))
    Model_svc= SVC(kernel=i)
    Model_svc.fit(X_train,Y_train)

    Train_score=Model_svc.score(X_train,Y_train)
    Test_score=Model_svc.score(X_test,Y_test)
    print('Train score for this kernel is {} \n Test score for this kernel is {}'.format(Train_score,Test_score))
    New_Pred=Model_svc.predict(X_test)
    Comp=({'Actual': Y_test, 'New_Prediction': New_Pred})
   
    print("CLASSIFICATION REPORT")
    Report = classification_report(Y_test,New_Pred)
    print(Report)
    print("CONFUSION MATRIX")
    cm =confusion_matrix(Y_test,New_Pred)
    sns.heatmap(cm,annot=True)
    plt.show()
```


### Using K-Nearest Neighbor
